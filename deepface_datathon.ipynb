```python

from deepface import DeepFace
import os as os
import csv
import matplotlib.pyplot as plt
```

```python

def get_age(img):
    age = -1
    try:
        result = DeepFace.analyze(img, actions = ['age'])[0]['age']
        age = result
    except: 
        print(f'I couldnt find the age for this {img}')

    return age

def get_gender(img):
    gender = 'unknown'
    try:
        results = DeepFace.analyze(img, actions = ['gender'])[0]['gender']
        gender = max(results, key=results.get)
    except:
        print(f'i couldnt find the gender for {img}')

    return gender


def get_race(img):
    race = 'unknown'
    try:
        results = DeepFace.analyze(img, actions = ['race'])[0]['race']
        race = max(results, key=results.get)
    except:
        print(f'i couldnt find the race for {img}')

    return race

def get_image_names(folderName):
    folder_path= folderName
    image_names = os.listdir(folder_path)
    image_names = [os.path.join(folder_path,x ) for x in image_names if x.endswith('.png')]
    return image_names




def create_csv(results):
    print(len(results))
    filename = r'C:\Users\Givan\Documents\ISL\results.csv'
    with open(filename, 'w'  ) as csvfile:
        csvfile.write('filename, gender, race, age\n')
        for r in results.values():
            csvfile.write(f"{r['filename']},{r['gender']},{r['race']},{r['age']}\n")
  



def main():
    
    print(os.getcwd())
    folderName = 'faceimages'
    images = get_image_names(folderName)
    results_dict = {}
    for image in images:
        age = get_age(image)
        gender = get_gender(image)
        race = get_race(image)
        results_dict[image] = {'filename': image, 'age': age, 'gender' : gender, 'race' : race }
    create_csv(results_dict)
    print(results_dict)

main()
```
